**EXTRACT AUDIO FILE FROM VIDEO**

```python
import subprocess

# Input video path
video_path = '/content/msd.mp4'

# Output audio path in MP3 format
audio_path = '/content/audio1.mp3'

try:
    # Run FFmpeg command to extract audio and convert it to MP3
    result = subprocess.run(
        [
            'ffmpeg',           # FFmpeg command
            '-i', video_path,   # Input video path
            '-vn',              # Disable video recording
            '-ac', '2',         # Set audio channels to stereo
            '-ar', '44100',     # Set audio sample rate to 44.1kHz
            '-ab', '192k',      # Set audio bit rate to 192 kbps
            '-f', 'mp3',        # Specify the output format as MP3
            audio_path           # Output audio path
        ],
        stdout=subprocess.PIPE,
        stderr=subprocess.PIPE,
        text=True
    )

    # Print FFmpeg's stderr output (which includes any errors or information)
    print(result.stderr)

except Exception as e:
    print(f"Error: {e}")
```

**TRANSLATE THE SUBTITLES**

```python
import pysrt
from googletrans import Translator

def translate_srt(input_file, output_file, target_language='ml'):
    # Load subtitles from the input file
    subtitles = pysrt.open(input_file, encoding='utf-8')

    # Initialize the Google Translator
    translator = Translator()

    # Translate each subtitle to the target language
    for subtitle in subtitles:
        # Translate only if the subtitle is in Hindi (assuming 'hi' for Hindi language code)
        if subtitle.text and subtitle.text != '॥':
            translation = translator.translate(subtitle.text, dest=target_language)
            subtitle.text = translation.text

    # Save the translated subtitles to the output file
    subtitles.save(output_file, encoding='utf-8')
if __name__ == "__main__":
    input_srt_file = "audio1.srt"
    output_srt_file = "Mal_subs.srt"

    translate_srt(input_srt_file, output_srt_file)
```

**SYNCING THE SUBTITILES WITH THE VIDEO**

```python
from moviepy.config import change_settings

# Specify the path to the ImageMagick binary in the Colab environment
change_settings({"IMAGEMAGICK_BINARY": "/usr/bin/convert"})

from moviepy.editor import VideoFileClip, TextClip, CompositeVideoClip

# Function to convert time format to seconds
def convert_time_to_seconds(timecode):
    h, m, s = map(float, timecode.replace(',', '.').split(':'))
    return h * 3600 + m * 60 + s

# Function to sync video and subtitle
def sync_video_subtitle(video_path, subtitle_path, output_path, start_time, end_time):
    # Load video clip
    video_clip = VideoFileClip(video_path)

    # Read subtitles from the subtitle file with proper encoding
    with open(subtitle_path, 'r', encoding='utf-8-sig') as file:
        lines = file.read().splitlines()

    # Initialize an empty list to store TextClips
    subtitle_clips = []

    # Process each subtitle entry
    for i in range(0, len(lines), 4):
        # Extract start and end time of the subtitle
        start, end = lines[i+1].split(' --> ')
        current_subtitle_start = convert_time_to_seconds(start)

        # Extract the subtitle text
        subtitle_text = lines[i+2]

        # Create a TextClip for the subtitle with a font that supports both English and Hindi
        subtitle_clip = TextClip(subtitle_text, fontsize=45, color='white', bg_color='black', size=(video_clip.size[0], None), font='/content/nirmala.ttf', method='caption')

        # Set the duration and position of the TextClip
        subtitle_clip = subtitle_clip.set_duration(convert_time_to_seconds(end) - current_subtitle_start).set_start(current_subtitle_start).set_pos(('center', 'bottom'))

        # Append the TextClip to the list
        subtitle_clips.append(subtitle_clip)

    # Create a CompositeVideoClip from the list of TextClips
    final_clip = CompositeVideoClip([video_clip] + subtitle_clips)

    # Write the final video with subtitles
    final_clip.write_videofile(output_path, codec='libx264', audio_codec='aac')

# Specify the paths
video_path = '/content/msd.mp4'
subtitle_path = '/content/Mal_subs.srt'
output_path = 'Mal_video.mp4'

# Call the sync_video_subtitle function with the uploaded file paths
sync_video_subtitle(video_path, subtitle_path, output_path, start_time=10, end_time=30)
```
